#### LSTM Modeling Cross Validation

```python
import pandas as pd
import numpy as np

#Validation Imports
from sklearn import metrics
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import StratifiedKFold

import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
from keras.models import load_model

#For clean Visualization
import warnings
warnings.filterwarnings('ignore')
```

## Importing clean dataset

```python
df = pd.read_csv("../Dataset/clean_data.csv")
df=df.dropna()
df.head()
```

```python
X=df.drop('label',axis=1)
Y=df['label']
```

## Vectorizing the Text

```python
voc_size=5000
sent_length = 20

seed = 7
np.random.seed(seed)
```

```python
onehot_repr=[one_hot(words,voc_size)for words in df['text'].values] 
```

```python
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)

X = np.array(embedded_docs)
```

## Modeling

Creating a cross validation instance, with 10 folds.

```python
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
cvscores = []
```

```python
embedding_vector_features = 40

for train, test in kfold.split(X, Y):
    print(len(train), len(test))

    X_train = X[train]
    Y_train = Y[train]

    model = Sequential()
    model.add(
        Embedding(voc_size,
                  embedding_vector_features,
                  input_length=sent_length))
    model.add(Dropout(0.3))
    model.add(LSTM(100))
    model.add(Dropout(0.3))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    model.fit(X_train, Y_train, epochs=50, batch_size=64, verbose=True)

    scores = model.evaluate(X[test], Y[test], verbose=True)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))
    cvscores.append(scores[1] * 100)

print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
```
